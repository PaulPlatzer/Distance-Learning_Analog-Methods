# Optimize distance : forecasting of tropical cylone data as a function of horizon

This notebook accompanies the following publication:
Paul Platzer, Arthur Avenas, Bertrand Chapron, Lucas Drumetz, Alexis Mouche, Léo Vinour. Distance Learning for Analog Methods. 2024. [⟨hal-04841334⟩](https://hal.science/hal-04841334)

It is used to plot the result of optimization algorithms for numerical experiments with IBTrACS tropical cyclone data, varying the forecast horizon.

```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import matplotlib
from tqdm.notebook import tqdm
from sklearn.neighbors import NearestNeighbors
from scipy.stats import bootstrap
import os
os.environ["OMP_NUM_THREADS"] = "16"
os.environ["OPENBLAS_NUM_THREADS"] = "16"
os.environ["MKL_NUM_THREADS"] = "16"
os.environ["NUMEXPR_NUM_THREADS"] = "16"
import sys
sys.path.append('../../functions/.')
from analogs import apply_transform, find_analogues, compute_weights, compute_diffs, compute_mae_mad, compute_error
from TC_utils import M, Rmax_from_M, correct_vmx_ibt, Rmxa23
```

```python
matplotlib.rcParams.update({'font.size': 14})
cols = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
```

```python
data_folder = '../../data/tropical_cyclone/'
output_folder = '../../output/tropical_cyclone/'
```

# Parameters for loading IBTrACS dataset

```python
files = os.listdir(data_folder)

# Input variables
var_names = ['Vmax', 'Rmax_IBT', 'R34', 'fcor', 'u_trans', 'v_trans']

# Output variable to forecast: derivative of Vmax
var_y = ['Vmax']
dydt = True
ind_var_y = []
for name_tmp in var_y:
    ind_var_y.append(np.argwhere(np.array(var_names)==name_tmp)[0][0])

# Utils to compute Rmax estimate from Avenas et al. (2023)
var_A23 = ['fcor', 'Vmax', 'R34', ]
ind_A23 = []
for name_tmp in var_A23:
    ind_A23.append(np.argwhere(np.array(var_names)==name_tmp)[0][0])

# Add names of auxilliary variables (Rmax_A23 and time-derivatives)
var_names_all = var_names.copy()
var_names_all.append('Rmax_A23')
for name in var_names_all.copy():
    var_names_all.append('d'+name+'/dt')

# Add name of time since the threshold of 18m/s is crossed for Vmax
var_names_all.append('t_18')
```

# Load optimization results

```python
# Results of gradient-descent optimization
npzfile = np.load(output_folder + 'optim_TC_horizon_gradient-descent.npz')
var_y = npzfile['var_y']
var_names_all = npzfile['var_names_all']
k = npzfile['k']
corr_length_train = npzfile['corr_length_train']
A_grad = npzfile['A']
# E_grad_train = npzfile['E_train']
E_grad_batch = npzfile['E_batch']
E_grad_test = npzfile['E_test']
E_unoptim_test = E_grad_test[:,:,0]
hh = npzfile['hh']
random_state_number = npzfile['random_state_number']
Nperm = len(random_state_number)
nn_algo = 'auto'

# Benchmarks and train-set error of gradient-descent optimization
npzfile = np.load(output_folder + 'benchmarks_for_plot_horizon.npz')
E_unoptim_train = npzfile['E_unoptim_train']
E_grad_train = npzfile['E_grad_train']
MAE_pers = npzfile['MAE_pers']
CRPS_clim = npzfile['CRPS_clim']
scale_dV = npzfile['scale_dV']

# Results of iterative grid-search optimization
npzfile = np.load(output_folder + 'optim_TC_horizon_grid_search_permut1.npz')
A_grid = npzfile['A_grid']
E_grid_train = npzfile['E_grid_train']
E_grid_test = npzfile['E_grid_test']
random_state_number_grid = npzfile['random_state_number']

for i in range(2,8+1):
    npzfile = np.load(output_folder + 'optim_TC_horizon_grid_search_permut'+str(i)+'.npz')
    A_grid = np.concatenate( ( A_grid , npzfile['A_grid'] ) , axis=1 )
    E_grid_train = np.concatenate( ( E_grid_train , npzfile['E_grid_train'] ) , axis=1 )
    E_grid_test = np.concatenate( ( E_grid_test , npzfile['E_grid_test'] ) , axis=1 )
    random_state_number_grid = np.concatenate( ( random_state_number_grid , npzfile['random_state_number'] ) )

Nperm_grid = len(random_state_number_grid)
```

# Plots

```python
gain_grad_train = 100*( E_unoptim_train - E_grad_train ) / E_unoptim_train
gain_grad_test = 100*( E_unoptim_test - E_grad_test[:,:,-1] ) / E_unoptim_test

gain_grid_train = 100*( E_unoptim_train[:,:Nperm_grid] - E_grid_train[:,:,-1] ) / E_unoptim_train[:,:Nperm_grid]
gain_grid_test = 100*( E_unoptim_test[:,:Nperm_grid] - E_grid_test[:,:,-1] ) / E_unoptim_test[:,:Nperm_grid]

gain_grad_vs_grid_train = 100*( E_grid_train[:,:,-1] - E_grad_train[:,:Nperm_grid] ) / E_grid_train[:,:,-1]
gain_grad_vs_grid_test = 100*( E_grid_test[:,:,-1] - E_grad_test[:,:Nperm_grid,-1] ) / E_grid_test[:,:,-1]
```

```python
av_test = np.zeros(len(hh))
av_test_CI = np.zeros( ( len(hh) , 2 ) )

av_train = np.zeros(len(hh))
av_train_CI = np.zeros( ( len(hh) , 2 ) )

for i_h in range(len(hh)):
    av_test[i_h] = np.mean(gain_grad_vs_grid_test[i_h])
    av_train[i_h] = np.mean(gain_grad_vs_grid_train[i_h])

    CI_test = bootstrap(gain_grad_vs_grid_test[i_h].reshape(1,-1), np.mean, ).confidence_interval
    av_test_CI[i_h] = CI_test.low, CI_test.high
    
    CI_train = bootstrap(gain_grad_vs_grid_train[i_h].reshape(1,-1), np.mean, ).confidence_interval
    av_train_CI[i_h] = CI_train.low, CI_train.high
```

```python
fig, ax = plt.subplots( 4 , 1 , figsize=(4,13) , sharex=True )

ax[0].plot( hh*3, np.mean( MAE_pers , axis=1 ) * scale_dV[:,0], ':', color=cols[0], label=r'persistence')
ax[0].plot( hh*3, np.mean( CRPS_clim , axis=1 ) * scale_dV[:,0], 'o-', color=cols[1], label=r'climatology')
ax[0].plot( hh*3, np.mean( E_unoptim_test , axis=1 ) * scale_dV[:,0], '-.', color=cols[2], label='unoptimized')
ax[0].plot( hh*3, np.mean( E_grid_test[:,:,-1] , axis=1 ) * scale_dV[:,0], '--', color=cols[4], label='grid-search (3 var.)')
ax[0].plot( hh*3, np.mean( E_grad_test[:,:,-1] , axis=1 ) * scale_dV[:,0], '-', color=cols[3], label='gradient-descent')
ax[0].set_xticks(hh[1::2]*3)
ax[0].set_xlim([12,120])
ax[0].legend( bbox_to_anchor=(1.01, 1), loc='upper left', ncol=1)
ax[0].set_ylabel(r'CRPS( $\Delta V_{max}$ )  (m s$^{-1}$)')
ax[0].grid()
ax[0].set_title('(a)')

ax[1].fill_between(hh*3, np.quantile(gain_grad_train, q=.25, axis=1),
                  np.quantile(gain_grad_train, q=.75, axis=1), color=cols[0], alpha=.5)
ax[1].plot(hh*3, np.median(gain_grad_train,axis=1), '-', color=cols[0], label='train')
ax[1].fill_between(hh*3, np.quantile(gain_grad_test, q=.25, axis=1),
                  np.quantile(gain_grad_test, q=.75, axis=1), color=cols[1], alpha=.3)
ax[1].plot(hh*3, np.median(gain_grad_test, axis=1), '--', color=cols[1], label='test')
ax[1].set_ylabel('CRPS loss (%) vs.\nunptimized')
ax[1].set_ylim([5,13]); 
ax[1].set_xticks(hh[1::2]*3)
ax[1].set_xlim([12,120])
ax[1].legend(); ax[1].grid()
ax[1].set_title('(b)')

for iplot in range(2):
    ax[2+iplot].plot(hh*3, np.zeros(len(hh)) , '--k')
    ax[2+iplot].grid()
    for i_perm in range(Nperm_grid):
        ax[2+iplot].scatter(hh*3, [gain_grad_vs_grid_train , gain_grad_vs_grid_test][iplot][:,i_perm], color='k', s=5, zorder=10)
    ax[2+iplot].plot(hh*3, [av_train, av_test][iplot], color='k')
    ax[2+iplot].fill_between(hh*3, [av_train_CI, av_test_CI][iplot][:,0], [av_train_CI, av_test_CI][iplot][:,1], color='k', alpha=.2)
    ax[2+iplot].set_ylabel('CRPS loss (%) vs.\ngrid-search (3 var.)')
    ax[2+iplot].text(75, 4.8, [r'$train$ $set$',r'$test$ $set$'][iplot])
    ax[2+iplot].set_ylim([-3,6]); ax[2+iplot].set_xlim([12,120])
    ax[2+iplot].set_xticks(hh[1::2]*3)
    ax[2+iplot].set_title(['(c)','(d)'][iplot])
ax[3].set_xlabel('forecast horizon (hours)')

plt.savefig(output_folder + 'fig_TC_optim_horizon.pdf', bbox_inches='tight', dpi=200)

plt.show()
```
