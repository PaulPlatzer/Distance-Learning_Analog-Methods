# Select variables : forecasting of tropical cylone data at 1-day horizon

This notebook accompanies the following publication:
Paul Platzer, Arthur Avenas, Bertrand Chapron, Lucas Drumetz, Alexis Mouche, Léo Vinour. Distance Learning for Analog Methods. 2024. [⟨hal-04841334⟩](https://hal.science/hal-04841334)

It is used to plot the results of optimization algorithms for numerical experiments with IBTrACS tropical cyclone data, at 1-day forecast horizon. The aim is to select the most relevant variables for this forecast and for this particular dataset.

```python
matplotlib.rcParams.update({'font.size': 14})
```

```python
cols = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
```

```python
data_folder = '../../data/tropical_cyclone/'
output_folder = '../../output/tropical_cyclone/'
```

# Parameters for loading IBTrACS dataset

```python
files = os.listdir(data_folder)

# Input variables
var_names = ['Vmax', 'Rmax_IBT', 'R34', 'fcor', 'u_trans', 'v_trans']

# Output variable to forecast: derivative of Vmax
var_y = ['Vmax']
dydt = True
ind_var_y = []
for name_tmp in var_y:
    ind_var_y.append(np.argwhere(np.array(var_names)==name_tmp)[0][0])

# Utils to compute Rmax estimate from Avenas et al. (2023)
var_A23 = ['fcor', 'Vmax', 'R34', ]
ind_A23 = []
for name_tmp in var_A23:
    ind_A23.append(np.argwhere(np.array(var_names)==name_tmp)[0][0])

# Add names of auxilliary variables (Rmax_A23 and time-derivatives)
var_names_all = var_names.copy()
var_names_all.append('Rmax_A23')
for name in var_names_all.copy():
    var_names_all.append('d'+name+'/dt')

# Add name of time since the threshold of 18m/s is crossed for Vmax
var_names_all.append('t_18')
```

# Loading and preprocessing dataset

```python
## Set forecast time-horizon (multiple of 3hours)

h = 8


## Load dataset

IBT = np.array(pandas.read_csv(data_folder + files[0], usecols = var_names))
IBT = np.concatenate( [ IBT , 
         Rmxa23(IBT[:,ind_A23[0]] , IBT[:,ind_A23[1]] , IBT[:,ind_A23[2]]).reshape(-1,1) ,
                      ],  axis=1)
IBT = np.concatenate( ( IBT[1:] , IBT[1:] - IBT[:-1] ) , axis=1 )
IBT = np.concatenate( [ IBT ,
           3*np.arange(len(IBT)).reshape(-1,1) ],  axis=1)
train_x = IBT[0:-h,:]
train_y = IBT[h:,ind_var_y] - IBT[0:-h,ind_var_y] 
ID = np.array([0]*len(IBT[0:-h,:]))


for i in np.arange(1, len(files)):
    IBT = np.array(pandas.read_csv(data_folder + files[i], usecols = var_names))
    IBT = np.concatenate( [ IBT , 
             Rmxa23(IBT[:,ind_A23[0]] , IBT[:,ind_A23[1]] , IBT[:,ind_A23[2]]).reshape(-1,1) ,
                          ],  axis=1)
    IBT = np.concatenate( ( IBT[1:] , IBT[1:] - IBT[:-1] ) , axis=1 )
    IBT = np.concatenate( [ IBT ,
           3*np.arange(len(IBT)).reshape(-1,1) ],  axis=1)
    train_x = np.concatenate([train_x, IBT[0:-h,:]])
    train_y = np.concatenate([train_y, IBT[h:,ind_var_y] - IBT[0:-h,ind_var_y]])
    ID = np.concatenate([ID, np.array([i]*len(IBT[0:-h,:]))])

# center and reduce
mean_IBTrACS = np.mean(train_x, axis=0)
std_IBTrACS = np.std(train_x, axis=0)
mean_y = np.mean(train_y, axis=0)
std_y = np.std(train_y, axis=0)
for j in range(train_x.shape[1]):
    train_x[:,j] = (train_x[:,j] - mean_IBTrACS[j]) / std_IBTrACS[j]
for j in range(train_y.shape[1]):
    train_y[:,j] = (train_y[:,j] - mean_y[j]) / std_y[j]    
```

# Loading optimization results

```python
# Load result of gradient-descent optimization
npzfile = np.load(output_folder + 'select_vars_TC_gradient_descent.npz')
var_y = npzfile['var_y']
var_names_all = npzfile['var_names_all']
A_grad = npzfile['A_grad']
E_grad_batch = npzfile['E_grad_batch']
# E_grad_train = npzfile['E_grad_train'] # -> only nans if not pre-computed
E_grad_test = npzfile['E_grad_test']
h = npzfile['h']
k = npzfile['k']
corr_length_train = npzfile['corr_length_train']
learning_rate_factor = npzfile['learning_rate_factor']
Regul = npzfile['Regul']
regul_type = npzfile['regul_type']
random_state_number = npzfile['random_state_number']
Nperm = A_grad.shape[0]

# Load pre-computed train-set error
npzfile = np.load(output_folder + 'train_error_for_plot_regularization.npz')
E_grad_train = npzfile['E_grad_train']

# Load result of iterative grid-search optimization
npzfile = np.load(output_folder + 'select_vars_TC_grid_search.npz')
A_grid = npzfile['A_grid']
E_grid_train = npzfile['E_grid_train']
E_grid_test = npzfile['E_grid_test']
```

# Plots

```python
plt.figure(figsize=(4,5),dpi=150)
for j in range(A_grad.shape[3]):
    if var_names_all[j][0] == 'd':
        ls = '--'
        color = cols[j % (len(var_names_all)//2) ]
    elif j == A_grad.shape[3]-1:
        ls = ':'
        color = 'k'
    else:
        ls = '-'
        color = cols[j]
#         plt.plot(A[0][0,k,:,i], ls=ls, color=color, label=var_names_all[i])
    plt.plot(Regul, np.mean(A_grad[:,:,-1,j], axis=0), ls = ls, color=color, label=var_names_all[j])
plt.yticks(np.arange(0,3,.5))
plt.xticks(Regul)
plt.legend(bbox_to_anchor = (1.02,0), loc = 3)
plt.yticks(np.arange(0,3,.5))
plt.ylabel('average weighting coefficient')
plt.ylim([0,3]); plt.xlim(Regul[0], Regul[-1])
plt.xlabel(r'regularization $\lambda$')
plt.title('horizon = '+str(int(3*h/24))+' day')
plt.savefig(output_folder + 'fig_TC_optim_grad_regul.pdf', bbox_inches='tight', dpi=200)
plt.show()
```

```python
av_train = np.zeros( len(Regul) )
av_train_CI = np.zeros( ( len(Regul) , 2 ) )
av_test = np.zeros( len(Regul) )
av_test_CI = np.zeros( ( len(Regul) , 2 ) )
for i_regul in range( len(Regul) ):    
    error_change_train = ( 100*(E_grid_train[:,-1] - E_grad_train[:,i_regul])/E_grid_train[:,-1] ).copy()
    error_change_test = ( 100*(E_grid_test[:,-1] - E_grad_test[:,i_regul,-1])/E_grid_test[:,-1] ).copy()
        
    av_train[i_regul] = np.mean(error_change_train)
    av_test[i_regul] = np.mean(error_change_test)
        
    CI_train = bootstrap(error_change_train.reshape(1,-1), np.mean).confidence_interval
    CI_test = bootstrap(error_change_test.reshape(1,-1), np.mean, ).confidence_interval
    av_train_CI[i_regul] = CI_train.low, CI_train.high
    av_test_CI[i_regul] = CI_test.low, CI_test.high
```

```python
thresh = .15
num_coef = np.sum( np.abs(A_grad[:,:,-1,:])>thresh , axis=-1 )
```

```python
# Change of error when switching from grid-search to gradient-descent
error_change_test_regul = 100*(E_grid_test[:,-1][:, np.newaxis] - E_grad_test[:,:,-1])/E_grid_test[:,-1][:, np.newaxis]
error_change_train_regul = 100*(E_grid_train[:,-1][:, np.newaxis] - E_grad_train[:,:])/E_grid_train[:,-1][:, np.newaxis]

# Same variable but indexed over constant number of significant coefficients
ncoef_axis = np.arange(14,2,-1)

error_change_test_ncoef = []
av_test_ncoef = np.zeros(len(ncoef_axis))
av_test_CI_ncoef = np.zeros( ( len(ncoef_axis) , 2 ) )

error_change_train_ncoef = []
av_train_ncoef = np.zeros(len(ncoef_axis))
av_train_CI_ncoef = np.zeros( ( len(ncoef_axis) , 2 ) )

for i_ncoef in range(len(ncoef_axis)):
    error_change_test_ncoef.append( error_change_test_regul[ num_coef == ncoef_axis[i_ncoef] ] )
    error_change_train_ncoef.append( error_change_train_regul[ num_coef == ncoef_axis[i_ncoef] ] )
    
    av_test_ncoef[i_ncoef] = np.mean(error_change_test_ncoef[-1])
    av_train_ncoef[i_ncoef] = np.mean(error_change_train_ncoef[-1])

    if np.sum(num_coef == ncoef_axis[i_ncoef]) > 1:
        CI_test = bootstrap(error_change_test_ncoef[-1].reshape(1,-1), np.mean, ).confidence_interval
        av_test_CI_ncoef[i_ncoef] = CI_test.low, CI_test.high
        CI_train = bootstrap(error_change_train_ncoef[-1].reshape(1,-1), np.mean, ).confidence_interval
        av_train_CI_ncoef[i_ncoef] = CI_train.low, CI_train.high
    else:
        av_test_CI_ncoef[i_ncoef] = np.nan
        av_train_CI_ncoef[i_ncoef] = np.nan
```

```python
plt.figure(figsize=(12,5))
for iplot in range(2):
    plt.subplot(121+iplot)
    plt.grid()
    plt.plot(ncoef_axis , np.zeros(len(ncoef_axis)) , '--k')
    for i_ncoef in range(len(ncoef_axis)):
        plt.scatter( [ ncoef_axis[i_ncoef] ] * len( [error_change_train_ncoef , error_change_test_ncoef][iplot][i_ncoef] ) ,
                    [error_change_train_ncoef , error_change_test_ncoef][iplot][i_ncoef] , color='k', s=1 , zorder=10 )
    plt.plot(ncoef_axis, [av_train_ncoef , av_test_ncoef][iplot], color='k')
    plt.fill_between(ncoef_axis, [av_train_CI_ncoef, av_test_CI_ncoef][iplot][:,0], [av_train_CI_ncoef, av_test_CI_ncoef][iplot][:,1], color='k', alpha=.2 )
    plt.ylim([-1.7,2.8])
    plt.xlabel('Number of coefficients')
    plt.ylabel(['CRPS loss (%) from algorithm\ncompared to grid-search (3 vars)', ''][iplot])
    plt.title(['(a) Train-set',
               '(b) Test-set'][iplot])
plt.savefig(output_folder + 'fig_TC_optim_grad_grid_ncoef.pdf', bbox_inches='tight', dpi=200)
plt.show()
```
